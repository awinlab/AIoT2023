## Step 1: Load data (also import library)

```python
# import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#import data
data=pd.read_csv("grade.csv")
test=pd.read_csv("test.csv")
print(data.head()) # show first 5 items
print(type(data))
print(data.info())


```

# Step 2: Pre-process X, Y
* 資料型別的 transformation) format transform (轉換成numpy format)
* (missing data=> imputation, 
* normalization
* data type 例如 categorical data onehot encoding, Label_Encoding, padas 套件有一個簡單的 get_dummies 的API



```python


X=data.iloc[:,:-1]
Y=data.iloc[:,-1]
print(X.columns)

# test data
testX=test.iloc[:,:-1]

X=X.values.reshape(-1,3)
Y=Y.values.reshape(-1,1)
testX=testX.values.reshape(-1,3)
#4. 資料型別的 transformation) format transform (轉換成numpy format)



print(type(X),X.shape)
print(type(Y),Y.shape)

```

# Step 3: Build Model for training

```python
from sklearn.linear_model import LogisticRegression as LR
model=LR()  # constructor
#all in policy
model.fit(X,Y) #

print(model) 

```

# Step 4: Evalute Model
* 2022.10.3 使用 k-fold cross validation
* [kaggle 範例](https://www.kaggle.com/code/jnikhilsai/cross-validation-with-linear-regression)





```python
# confusion matrix
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score
# from sklearn.model_selection import KFold
# from sklearn.metrics import accuracy_score

score = cross_val_score(model, X, Y, cv=5).mean()
print(score)

preY=model.predict(X)
testY=model.predict(testX)
print("confusion matrix=")
print(confusion_matrix(Y, preY))

# Accuracy only




# 高階API classification report
# from sklearn.metrics import classification_report 
# target_names = ['fail', 'pass']
# print(classification_report(Y, preY, target_names=target_names))

```

# Step 5: Deploy Model to predict new value

```python

testY=model.predict(testX)
test['pass']=testY

test.to_csv("result.csv", index=False, mode='w')
```
